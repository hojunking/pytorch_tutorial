```python
import albumentations as A
import cv2
import matplotlib.pyplot as plt

image1 = cv2.imread("../efficientnet_pytorch/test_data/_2019.jpg")
image2 = cv2.imread("../efficientnet_pytorch/test_data/avan2.jpg")
height = 300
width = 300

image = image1
# Declare an augmentation pipeline
transform = A.Compose([
     A.Resize(height = 380, width = 380), 
#    A.CenterCrop(always_apply=True,p=1.0, height=380, width=380),
#     A.RandomCrop(always_apply=True, p=1.0, height=380, width=380),
    A.RandomResizedCrop(
    height = 380,
    width = 380,
    scale=(0.40, 0.46),
    ratio=(0.90, 1.10),
     always_apply=True),
#    A.HorizontalFlip(p=0.5),
#    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225], max_pixel_value=255.0, p=1.0)
    #A.pytorch.transforms.ToTensorV2()
])

augmentations = transform(image=image)
augmentation_img = augmentations["image"]

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.grid(False)
plt.imshow(image, cmap=plt.cm.binary)

plt.subplot(2,2,2)
plt.imshow(augmentation_img, cmap=plt.cm.binary)
plt.show()
```
