```python
import albumentations as A
import cv2

image1 = cv2.imread("../Data/test/1.jpg")
image2 = cv2.imread("../Data/test/4.jpg")

height = 600
width = 600

image = image1
# Declare an augmentation pipeline
transform = A.Compose([
    #A.Resize(height=height, width=width),
    A.RandomResizedCrop(
            height=600, 
            width=600, 
            scale=(0.24, 0.26),
            ratio=(0.90, 1.10),
            always_apply=True
            ),
    A.HorizontalFlip(p=0.5),
    A.Normalize(mean=(0.548, 0.504, 0.479), std=(0.237, 0.247, 0.246))
    #A.pytorch.transforms.ToTensorV2()
])

augmentations = transform(image=image)
augmentation_img = augmentations["image"]

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.grid(False)
plt.imshow(image, cmap=plt.cm.binary)

plt.subplot(2,2,2)
plt.imshow(augmentation_img, cmap=plt.cm.binary)
plt.show()
```
