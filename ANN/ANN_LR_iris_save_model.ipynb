```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Model(nn.Module):

    def __init__(self, in_features =4, h1 =8, h2=9, out_features=3):
        # How many layers?
        super().__init__()
        self.fc1 = nn.Linear(in_features, h1)
        self.fc2 = nn.Linear(h1,h2)
        self.out = nn.Linear(h2,out_features)
        # layers (4 features) --> (h1 N) input --> (h2 N) --> (h3 N) output (3 classes)
    def forward(self,x):
        # activation functions
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        # output
        x = self.out(x)
        return x
```

```python
torch.manual_seed(32)
model = Model()
```

```python
# dataload iris dataset
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../PYTORCH_NOTEBOOKS/Data/iris.csv")
```

```python
df.head()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,7))
fig.tight_layout()

plots = [(0,1),(2,3),(0,2),(1,3)]
colors = ['b', 'r', 'g']
labels = ['Iris setosa','Iris virginica','Iris versicolor']

for i, ax in enumerate(axes.flat):
    for j in range(3):
        x = df.columns[plots[i][0]]
        y = df.columns[plots[i][1]]
        ax.scatter(df[df['target']==j][x], df[df['target']==j][y], color=colors[j])
        ax.set(xlabel=x, ylabel=y)

fig.legend(labels=labels, loc=3, bbox_to_anchor=(1.0,0.85))
plt.show()
```

```python
X = df.drop('target', axis = 1)
y = df['target']
```

```python
X = X.values
y = y.values
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state=33) ##shuffle\ out data
```

```python
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
## multi class -> one-hot-encoding requied but, if you use the cross entropy loss func, you can use it later
```

```python
y_train = torch.LongTensor(y_train)
```

```python
y_test = torch.LongTensor(y_test)
```

```python
criterion = nn.CrossEntropyLoss() ## use it when multi class
optimizer = torch.optim.Adam(model.parameters(),lr = 0.01) ##if the errors are not going down, then low lr recommend
```

```python
model.parameters
```

```python
## train
## How many EPOCHS? starts
# EPOCH 1 run via all the training data

epochs = 100
losses = []

for i in range(epochs):
    #Forward and get a prediction
    
    y_pred = model.forward(X_train)

    #calculate loss/error
    loss = criterion(y_pred, y_train) ## don't need to encode one-hot-encode
    losses.append(loss)

    if i % 10==0: ## 10 epoch print 
        print(f'epoch {i} and loss is: {loss}')

    #Backprop
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
```

```python
# losses = tensor.detach().numpy()
# plt.plot(range(epochs), losses)
# plt.ylabel('LOSS')
# plt.xlabel('EPOCHS')
```

```python
## evaluate metric
with torch.no_grad():
    y_eval = model.forward(X_test)
    loss = criterion(y_eval,y_test)
```

```python
loss ## if you ovefit to the trainig data, sort of huge loss
```

```python
correct = 0

with torch.no_grad():

    for i,data in enumerate(X_test):

        y_val = model.forward(data)

        print(f'{i+1}.) {str(y_val)} {y_test[i]}')

        if y_val.argmax().item() == y_test[i]:
            correct += 1


print(f'We got {correct} correct!!')
```
